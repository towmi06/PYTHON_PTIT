```python
from math import  *

class Phanso:
    def __init__(self, tu, mau):
        self.tu = tu
        self.mau = mau

    def toigian(self):
        ucln = gcd(self.tu , self.mau)
        self.tu //= ucln
        self.mau //= ucln

    def __str__ (self):
        return f'{self.tu}/{self.mau}'

if __name__ == '__main__':
    tu, mau = map(int,input().split())
    p = Phanso(tu,mau)
    p.toigian()
    print(p)
```

```python
from math  import*
class Gvien:
    def __init__(self, mgv, ten, lcb):
        self.__mgv = mgv
        self.__ten = ten
        self.__lcb = lcb
        self.__hso = int(mgv[-2:])  # Lấy hệ số là 2 ký tự cuối cùng của mã GV
        self.__luong = 0

    def tinh_luong(self):
        if self.__mgv[:2] == 'HT':
            self.__luong = self.__lcb * self.__hso + 2000000
        elif self.__mgv[:2] == 'HP':
            self.__luong = self.__lcb * self.__hso + 900000
        elif self.__mgv[:2] == 'GV':
            self.__luong = self.__lcb * self.__hso + 500000

    def __str__(self):
        return f'{self.__mgv} {self.__ten} {self.__hso} {self.__luong}'

if __name__ == '__main__':
    mgv = input()
    ten = input()
    lcb = int(input())
    gv = Gvien(mgv, ten, lcb)
    gv.tinh_luong()
    print(gv)
```

```python
class Nvien:
    # Từ điển chứa phụ cấp theo chức vụ
    phucap = {'GD': 250000, 'PGD': 200000, 'TP': 180000, 'NV': 150000}

    def __init__(self, mnv, ten, lcb, cong, cvu):
        self.mnv = mnv
        self.ten = ten
        self.lcb = lcb
        self.cong = cong
        self.cvu = cvu

    # Tính lương tháng
    def get_luongthang(self):
        return self.cong * self.lcb

    # Tính thưởng
    def get_thuong(self):
        luong = self.get_luongthang()
        if self.cong >= 25:
            return int(luong * 0.2)
        elif 22 <= self.cong < 25:
            return int(luong * 0.1)
        else:
            return 0

    # Tính phụ cấp dựa trên chức vụ. Nếu chức vụ không tồn tại, trả về 0.
    def get_phucap(self):
        return self.phucap.get(self.cvu, 0)

    # Tổng thu nhập
    def get_thunhap(self):
        return self.get_luongthang() + self.get_thuong() + self.get_phucap()

    def __str__(self):
        return f'{self.mnv} {self.ten} {self.get_luongthang()} {self.get_thuong()} {self.get_phucap()} {self.get_thunhap()}'


if __name__ == '__main__':
    ten = input()
    lcb = int(input())
    cong = int(input())
    cvu = input()
    nv = Nvien('NV01', ten, lcb, cong, cvu)
    print(nv)
```

```python
from math import*

class Product:
    cnt = 0
    def __init__(self,ten,dv, mua, ban):
        Product.cnt +=1
        self.ma = f'MH{Product.cnt:04d}'
        self.ten = ten 
        self.dv = dv
        self.mua = mua
        self.ban = ban

    def profit(self):
        return self.ban - self.mua
    def __str__(self):
        return f'{self.ma} {self.ten} {self.dv} {self.mua} {self.ban} {self.profit()}'

if __name__ == '__main__':
    n = int(input())
    a= []
    for _ in range(n):
        mh = Product(input(), input(), int(input()), int(input()))
        a.append(mh)
    
    a.sort(key=lambda x: (-x.profit(), x.ma))
    for x in a:
        print(x)
```

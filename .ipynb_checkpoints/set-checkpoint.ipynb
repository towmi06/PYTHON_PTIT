```python
from collections import Counter
from math import isqrt
import sys

# HÃ m kiá»ƒm tra sá»‘ nguyÃªn tá»‘
def is_prime(n):
    if n < 2:
        return False
    if n in (2, 3):
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    for i in range(5, isqrt(n) + 1, 6):  # Kiá»ƒm tra cÃ¡c sá»‘ dáº¡ng 6k Â± 1
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True

'''HÃ m is_prime Ä‘Ã£ Ä‘Æ°á»£c tá»‘i Æ°u hÃ³a báº±ng cÃ¡ch:
Tráº£ vá» False cho táº¥t cáº£ sá»‘ nhá» hÆ¡n 2 vÃ  True cho 2 vÃ  3.
Bá» qua táº¥t cáº£ cÃ¡c sá»‘ cháºµn vÃ  cÃ¡c bá»™i cá»§a 3, chá»‰ kiá»ƒm tra cÃ¡c sá»‘ dáº¡ng 
6ð‘˜ Â± 1. Äiá»u nÃ y giáº£m Ä‘Ã¡ng ká»ƒ sá»‘ láº§n cáº§n kiá»ƒm tra.
'''
if __name__ == '__main__':
    # Äá»c táº¥t cáº£ cÃ¡c sá»‘ tá»« Ä‘áº§u vÃ o
    a = list(map(int, sys.stdin.read().split()))
    
    # Sá»­ dá»¥ng Counter Ä‘á»ƒ Ä‘áº¿m táº§n suáº¥t
    cnt = Counter()
    
    # Äáº¿m sá»‘ láº§n xuáº¥t hiá»‡n cá»§a sá»‘ nguyÃªn tá»‘
    for x in a:
        if is_prime(x):
            cnt[x] += 1  # TÄƒng táº§n suáº¥t cá»§a sá»‘ nguyÃªn tá»‘

    # In ra káº¿t quáº£ theo thá»© tá»± xuáº¥t hiá»‡n ban Ä‘áº§u
    for key in cnt:
        print(key, cnt[key])
```

```python
#liet ke va dem , dem so khong giam 
import sys
from math import*
def check(n):
    r = n%10
    n // =10
    while n != 0:
        if r < n% 10:
            return 0
        r = n %10
        n //=10
    return 1
if __name__ == '__main__':
    d= dict({})
    a = list(map(int, sys.stdin.read().split()))
    for x in a:
        if check(x):
            if x in d:
                d[x] +=1
            else: d[x] = 1
    d = list(d.items())
    # Sáº¯p xáº¿p theo táº§n suáº¥t giáº£m dáº§n, sau Ä‘Ã³ theo giÃ¡ trá»‹ tÄƒng dáº§n
    d.sort(key = lambda x: (-x[1], x[0]))
    for x, y in d:
        print(x,y)
```

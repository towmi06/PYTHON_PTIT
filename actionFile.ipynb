```python
# loc ra so nguyen to va theo thu tu tangdan 
from math import isqrt

def nto(n):
    if n < 2:
        return False
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

# S dng with  x l tp
with open('input3.txt', 'r') as f1, open('output3.txt', 'w') as f2:
    # c d liu t file v chuyn thnh danh sch s nguyn
    data = f1.read()
    a = list(map(int, data.split()))
    
    # Lc s nguyn t v sp xp theo th t tng dn
    b = sorted(x for x in a if nto(x))
    
    # Ghi danh sch s nguyn t vo tp output
    f2.write(' '.join(map(str, b)))

    
```

```python
#cap email 

from math import*
def email (s):
    a = s.lower().split()
    res = a[-1]
    for x in a[:-1]:
        res+=x[0]
    res += '@stu.ptit.edu.vn'
    return res

def mk (s):
    a = s.split('/')
    res = ''
    for x in a:
        res += str(int(x))
    return res

with open('input4.txt','r') as f1, open('output4.txt','w') as f2:

    data = []
    
    for line in f1:
        data.append(line[:-1])
    for i in range(0,len(data),2):
        f2.write(email(data[i]) + '\n')
        f2.write(mk(data[i+1]) +'\n')
        
```

```python
# lco nhung ban  co tong diem >= 27 va theo thu tu tangf dan 
class sinhvien:
    def __init__(self, ten, email, sdt, toan, li, hoa):
        self.ten = ten 
        self.email = email
        self.sdt = sdt
        self.toan = toan 
        self.li = li
        self.hoa = hoa

    def diem(self):
        return self.toan + self.li + self.hoa

    def __str__(self):
        return f'{self.ten}\n{self.email}\n{self.sdt}\n{self.diem()}\n'

# c d liu v ghi ra tp
with open('input5.txt', 'r') as f1, open('output5.txt', 'w') as f2:
    a = []
    data = [line.strip() for line in f1]

    # X l d liu u vo
    for i in range(0, len(data), 4):
        ten = data[i]
        email = data[i + 1]
        sdt = data[i + 2]
        t, l, h = map(int, data[i + 3].split())
        s = sinhvien(ten, email, sdt, t, l, h)
        a.append(s)

    # Lc v sp xp sinh vin
    f = sorted((sv for sv in a if sv.diem() >= 27), key=lambda x: x.diem())
    b = f[::-1]
    # Ghi danh sch sinh vin tha mn iu kin ra tp
    for sv in b:
        f2.write(str(sv))
```
